```python
import pandas as pd
import sentencepiece as spm
```

```python
df = pd.read_json('output\General_train_text_chunk_1_cleaned.jsonl', lines = True)
```

```python
# df['cleaned_abstract'] = df['cleaned_abstract'].apply(lambda x: 'sostok ' + x + ' eostok')
```

```python
df.columns
```

```python
spm.SentencePieceTrainer.Train(input = ['output\General_train_text_chunk_1_cleaned.jsonl','output\General_train_text_chunk_2_cleaned.jsonl','output\General_train_text_chunk_3_cleaned.jsonl','output\General_train_text_chunk_4_cleaned.jsonl','output\General_train_text_chunk_5_cleaned.jsonl'], model_prefix = 'bpe', vocab_size = 6000)
```

```python
sp = spm.SentencePieceProcessor(model_file = 'bpe.model')
```

```python
for i in range(len(df)):
    text = sp.encode(df.at[i, 'cleaned_text'])
    abstract = sp.encode(df.at[i, 'cleaned_abstract'])
    df.at[i, 'text_token_length'] = int(len(text))
    df.at[i, 'abstract_token_length'] = int(len(abstract))
```

```python
df
```

```python
print(min(df['text_token_length']), min(df['abstract_token_length']))
```

```python
df_filterd = df[
    (df['text_token_length'] <= 2500) &
    (df['text_token_length'] >= 250) & 
    (df['abstract_token_length'] <= 140) &
    (df['abstract_token_length'] >= 30)
]
df_filterd
```

```python
print(min(df_filterd['text_token_length']), min(df_filterd['abstract_token_length']))
```

```python
df_filterd.reset_index(drop = True)
```

```python
df_filterd.columns
```

```python
text_tokens = []
abstract_tokens = []

for tokens in range(len(df_filterd)):
    text = sp.encode(df_filterd.iloc[tokens]['cleaned_text'])
    abstract = sp.encode(df_filterd.iloc[tokens]['cleaned_abstract'])
    text_tokens.append(text)
    abstract_tokens.append(abstract)
```

```python
df_filterd.to_json('output\General_train_text_chunk_5_cleaned.jsonl', orient = 'records', lines = True)
df_tokens = pd.DataFrame(list(zip(text_tokens, abstract_tokens)), columns = ['text_tokens', 'abstract_tokens'])
df_tokens
```

```python
df_tokens.to_json('output/train_tokens_5.jsonl', orient = 'records', lines = True)
```
